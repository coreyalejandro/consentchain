# ITAYN Colab + Gradio Demo

This notebook runs the ITAYN MVP and exposes a simple Gradio UI.

```python
# Clone the repo and build the ITAYN package
!git clone https://github.com/coreyalejandro/COL-REASONS-ASF-2026.git
%cd COL-REASONS-ASF-2026/dev/anthropic-ai-safety-fellows-application-prompt/research/ITAYN
!npm install
!npm run build
```

```python
import subprocess
import gradio as gr

def run_itayn(text):
    script = (
        "const { createITAYNProcessor } = require('./dist/index.js');"
        "(async () => {"
        "  const processor = createITAYNProcessor();"
        "  const result = await processor.process({ content: process.argv[1], timestamp: new Date(), sessionId: 'colab' });"
        "  console.log(JSON.stringify(result, null, 2));"
        "})();"
    )
    cmd = ['node', '-e', script, text]
    output = subprocess.check_output(cmd)
    return output.decode('utf-8')

demo = gr.Interface(
    fn=run_itayn,
    inputs=gr.Textbox(lines=6, label='Input'),
    outputs=gr.Code(label='ITAYN Output (JSON)'),
    title='ITAYN MVP Demo',
    description='Heuristic-free output from the ITAYN pipeline.'
)
demo.launch(share=True)
```
